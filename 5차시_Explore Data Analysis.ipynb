```python
# 날짜 Type 변경
transactions['created_at'] = pd.to_datetime(transactions['created_at'], errors='coerce')
transactions['shipment_date_limit'] = pd.to_datetime(transactions['shipment_date_limit'], errors='coerce')
customer['birthdate'] = pd.to_datetime(customer['birthdate'], errors='coerce')
click_stream['event_time'] = pd.to_datetime(click_stream['event_time'], errors='coerce')
```

```python
#customer.info()
#transactions.info()
click_stream.info()
```

```python
# 연말 시즌 필터링 (최근 3개년)

# 3개년 계산
max_year = pd.to_datetime(transactions['created_at']).max().year - 1  #2021
years = [max_year - i for i in range(3)]

# 각 연도별 필터링
filtered_transactions = pd.DataFrame()
filtered_click_stream = pd.DataFrame()

# 각 연도별 11~12월 필터링
for y in years:
    start_date = f"{y}-11-01"
    end_date = f"{y}-12-31"

    t_filtered = transactions[
        (transactions['created_at'] >= start_date) &
        (transactions['created_at'] <= end_date)
    ]
    c_filtered = click_stream[
        (click_stream['event_time'] >= start_date) &
        (click_stream['event_time'] <= end_date)
    ]

    filtered_transactions = pd.concat([filtered_transactions, t_filtered])
    filtered_click_stream = pd.concat([filtered_click_stream, c_filtered])
```

```python
# 결과 요약 출력
print("기준 연도:", years)
print("분석 기간: 각 연도 11/01 ~ 12/31")
print("거래 수:", len(filtered_transactions))
print("클릭 로그 수:", len(filtered_click_stream))
```

```python
#결측치 처리
filtered_transactions['promo_amount'] = filtered_transactions['promo_amount'].fillna(0)
filtered_transactions['promo_code'] = filtered_transactions['promo_code'].fillna('None')
filtered_click_stream['traffic_source'] = filtered_click_stream['traffic_source'].fillna('Unknown')
```

```python
filtered_transactions.info()
#click_stream.info()
```

```python
# 연-월 컬럼 추가 (예: 2023-11)
filtered_transactions['year_month'] = filtered_transactions['created_at'].dt.to_period('M')

# 연월별 거래량과 매출 합계 집계
monthly_stats = (filtered_transactions.groupby('year_month').agg(
    transaction_count=('created_at', 'count'),  # 거래 건수
    sales_sum=('total_amount', 'sum')                 # 매출 합계
    ).reset_index()
)

# 결과 확인
display(monthly_stats)
```

```python
#이상치 처리
#나이 이상치 확인
customer['age'] = 2021 - customer['birthdate'].dt.year  #거래 기준 max날짜 기준
#display(customer)
print(customer['age'].max())
print(customer['age'].min())    #5 ->10
customer.loc[(customer['age'] < 10)] = np.nan
```

```python
filtered_transactions['promo_used'] = (filtered_transactions['promo_code'].str.lower() != 'none') | (filtered_transactions['promo_amount'] > 0)
filtered_transactions['is_success'] = (filtered_transactions['payment_status'] == 'Success').astype(int)
filtered_transactions['total_non_promo'] = filtered_transactions['total_amount'] - filtered_transactions['promo_amount']
```

```python
filtered_transactions.to_csv("output/transactions.csv", index=False, encoding='utf-8-sig')
filtered_click_stream.to_csv("output/click_stream.csv", index=False, encoding='utf-8-sig')
customer.to_csv("output/customer.csv", index=False, encoding='utf-8-sig')
product.to_csv("output/product.csv", index=False, encoding='utf-8-sig')
```

```python
transactions = filtered_transactions[filtered_transactions['payment_status'] == 'Success'].copy()

# 연령/연령대
anchor_year = int(transactions['created_at'].dt.year.max()) if transactions['created_at'].notna().any() else 2021
customer['age'] = anchor_year - customer['birthdate'].dt.year
customer.loc[(customer['age'] < 10) | (customer['age'] > 100), 'age'] = np.nan

bins = [10,20,30,40,50,60,200]
labels = ['10대','20대','30대','40대','50대','60대+']
customer['age_group'] = pd.cut(customer['age'], bins=bins, labels=labels, right=False)

# -------------------------
# 3) 조인
# -------------------------
merged = transactions.merge(customer[['customer_id','age_group']], on='customer_id', how='left')

# booking_id 보정 (없다면 대체)
if 'booking_id' not in merged.columns:
    merged['booking_id'] = merged.index.astype(str)

# -------------------------
# 4) 구매 기여도 계산 (전체 모수 기준)
#    y = (연령대별 promo_used=True 구매 수) / (전체 성공 구매 수) * 100
# -------------------------
total_success = merged['booking_id'].nunique()

pivot_cnt = (merged.groupby(['age_group','promo_used'])['booking_id']
             .nunique()
             .unstack(fill_value=0)
             .rename(columns={False:'NotPromo', True:'Promo'}))

pivot_cnt['TotalSuccess_AllAges'] = total_success
pivot_cnt['Promo_Contribution_Total(%)'] = (pivot_cnt['Promo'] / total_success * 100).round(2)

# (참고) 연령대 내부 프로모션 비중(의존도)
pivot_cnt['Promo_Dependency_within_Age(%)'] = (
    pivot_cnt['Promo'] / (pivot_cnt['Promo'] + pivot_cnt['NotPromo']).replace(0, np.nan) * 100
).round(2)

result = pivot_cnt.reset_index().sort_values('Promo_Contribution_Total(%)', ascending=False)
print("=== 연령대별 프로모션 구매 기여도 (전체 모수 기준) ===")
print(result[['age_group','Promo','NotPromo','TotalSuccess_AllAges','Promo_Contribution_Total(%)','Promo_Dependency_within_Age(%)']])
```
